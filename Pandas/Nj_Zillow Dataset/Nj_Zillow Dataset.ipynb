Single-family home prices of cities in the United States (City_SingleFamily.csv), is derrived from the Zillow Research (https://www.zillow.com/research/data/). Well perform some statistics on the average prices of the single-family home of cities in the United State from April, 1996 to October, 2018.  



```python
# Don't delete this cell
# Execute this cell to import pandas, numpy, and matplotlib.pyplot.

import pandas as pd
import numpy as nps
import matplotlib.pyplot as plt
pd.options.display.max_rows = 10
```


Lets read the data into python.

```python
country = pd.read_csv("City_SingleFamily.csv",thousands=',', index_col="RegionName")
country
```

Lets select all data concerning New Jersey. 

```python
newJersey = country.loc[country['State'] == 'NJ'].reset_index()
newJersey
```

Here ill perform some aggregation. Lets group **newJersey** by the county name and set the new values to be the average prices. New groupeddataset as **county**.


```python
county = newJersey.groupby('CountyName').mean()
county
#county.mean()
```


Lets indicate the data set **county** as time series.
  

```python
from datetime import datetime
####Try 1
county.reset_index()
#county = county.T
#county = pd.to_datetime(CountyName)
#county.set_index('CountyName', inplace=True)

####Try 2
#county = county.set_index('CountyName').T

####Try 3
county = county.T
county.index = pd.to_datetime(county.index)
#county.set_index('CountyName', inplace = True)

county
```


Lets do some resampling. We will create a new dataFrame called **county_year** which is the downsample dataset from **county** by setting the freqency as years and setting the new values to be the average prices. Both the closed side and the label should are 'left' for the downsample data. 

```python
#class 23: slide 22/35
#county_year = county.resample('years' ,closed='left', label='left').mean()
county_year = county.resample('1Y' ,closed='left', label='left').mean()
county_year
```
