This dataset(olympics.csv), is sourced from: from the Wikipedia entry on All Time Olympic Games Medals, and does some basic data cleaning. We perform some cleaning and statistics on the number of Summer/Winter Olympics each country has attended and the number of medals it has won.  



```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.max_rows = 10


olympics = pd.read_csv("olympics.csv",thousands=',', index_col="Country")
olympics
```


Lets Visualize the number of medals of top 10 Countries of the total number of medals in the summer olympics by a clustered bar plot.
  
 Well perform the following transformations:
1. Sort the olympics dataset by the values in the column 'SummerTotal' in descending order.
2. Select the first 10 rows of the sorted dataset and three columns: **SummerGold**, **SummerSilver**, and **SummerBronze**. Assign these 10 rows with three columns to a new DataFrame  **summer**.
3. Apply 'plot.bar()' to plot a clustered bar chart with title "Medals in Summer Olympics". The Axis-x shows different countries and the Axis-y indicates the number of medals. The lengend shows the three kinds of medals: "SummerGold", "SummerSilver", and "SummerBronze". The ticks in Axis-x should be rotated by 30 degree.


```python
summer = olympics.sort_values(by= 'SummerTotal', ascending=False)
#olympics = olympics.head(10)
summer = summer[['SummerGold', 'SummerSilver', 'SummerBronze']].head(10).reset_index()
#summer = summer.sort_values(by= 'SummerTotal', ascending=False)
#summer = olympics.loc[:, ('SummerGold', 'SummerSilver', 'SummerBronze')].copy()
#summer = summer.set_index('Country')
summer.plot.bar(rot=30, title="Medals in Summer Olympics")
#summer
```

Lets visualize the distribution of Gold, Silver, and Bronze medals in the summer Olympics for the top 10 countries by the box plot using Seaborn. The Axis-x indicates different kinds of medals and axis-y shows to number of medals. 
  
Steps:   
1. Create a new DataFrame called **melted** which is a copy of **summer**.
2. Set the index of DataFrame **melted** to be one column **Country**.
3. Apply 'melt' function in pandas to convert **melted** from 'wide' format to 'long' format.
3. Apply 'factorplot' function in Seaborn.

```python
###see class 19_dataVis.2
melted = summer.copy()
###indexing
#melted.set_index('Country', inplace = True)
#melted = melted.set_index('Country')
#melted = melted.reset_index()
melted = melted.set_index('Country').reset_index()

###MELT
#melted.melt(id_vars=['Country'], value_vars=['SummerGold', 'SummerSilver', 'SummerBronze'])
#melted.reindex(['UnitedStates', 'GreatBritain', 'SovietUnion', 'France', 'Germany', 'Italy', 'China', 'Australia', 'Sweden', 'Hungary'])
melted = pd.melt(melted, id_vars=['Country'], value_vars=['SummerGold', 'SummerSilver', 'SummerBronze'])

##factorplot renamed catplot
#sns.catplot(x="Country", y="value", data=melted)
sns.factorplot(x="variable", y="value", data=melted, kind = 'box')
#melted
```
